# Analisis Data Realisasi Belanja Sektor Ekonomi, Pendidikan dan Kesehatan serta Indeks Pembangunan Manusia

## Deskripsi
<b>Indeks Pembangunan Manusia</b> merupakan indeks yang mengukur pembangunan manusia dari tiga aspek dasar. Tiga aspek tersebut adalah 1) pengetahuan; 2) umur panjang, hidup sehat; dan 3) standar hidup layak.

<b>Pengeluaran pemerintah</b> (<i>government   expenditure</i>) adalah bagian dari kebijakan fiskal (Sukirno,2000) yakni suatu tindakan  pemerintah untuk mengatur jalannya perekonomian dengan cara menentukan besarnya penerimaan dan pengeluaran pemerintah tiap tahunnya yang tercermin dalam dokumen APBN untuk nasional dan APBD untuk daerah/regional. Tujuan dari kebijakan fiskal ini adalah dalam rangka menstabilkan harga, tingkat output maupun kesempatan kerja dan memacu pertumbuhan ekonomi.

Penelitian ini mencoba menganalisis hubungan antara belanja pemerintah daerah pada sektor ekonomi, pendidikan, dan kesehatan terhadap tingkat Indeks Pembangunan Manusia Indonesia.

Data Indeks Pembangunan Manusia diperoleh dari https://data.go.id/dataset/indeks-pembangunan-manusia-ipm-per-provinsi yang telah melalui proses penyiapan data agar siap untuk dilakukan analisis.
Data belanja pemerintah diperoleh dari https://data.go.id/dataset/indo-dapoer yang telah melalui proses penyiapan data agar siap untuk dilakukan analisis.

## Loading Dataset

```python
import pandas as pd
```

```python
belanja = pd.read_csv('INDODAPOERData.csv', delimiter = ',', encoding = "utf-8")
ipm = pd.read_csv('processed-ipm-per-provinsi-1996-2010.csv', delimiter = ',', encoding = "utf-8")
```

```python
belanja.dropna().sample(5)
```

```python
ipm.dropna().sample(5)
```

## Statistik Deskriptif  Data <i> Expenditure </i>

```python
belanja.groupby('Tahun').describe()['Ekonomi']
```

```python
belanja.groupby('Tahun').describe()['Pendidikan']
```

```python
belanja.groupby('Tahun').describe()['Kesehatan']
```

## Statistik Deskriptif Data IPM

```python
ipm.groupby('Tahun').describe()['IPM']
```

## Menggabungkan Data Menjadi Satu <i>Data Frame</i>

```python
data_gabung = belanja.merge(ipm, how='inner', on=['Kode_Provinsi', 'Tahun']).dropna()
data_gabung.sample(5)
```

## Statistik Yang Telah Digabungkan

```python
data_gabung.groupby('Tahun').describe()[['Ekonomi', 'Pendidikan', 'Kesehatan', 'IPM']]
```

## Korelasi Antara Education and Health Expenditure terhadap IPM

```python
vis = data_gabung.groupby('Tahun')[['Ekonomi', 'Pendidikan', 'Kesehatan', 'IPM']].mean()
vis.corr()
```

## Visualisasi Data

```python
import seaborn as sea

sea.lmplot(x = 'Ekonomi', y = 'IPM', data = vis)
```

```python
sea.lmplot(x = 'Pendidikan', y = 'IPM', data = vis)
```

```python
sea.lmplot(x = 'Kesehatan', y = 'IPM', data = vis)
```

## Analisis Dan Simpulan

Hasil penelitian di atas menunjukkan bahwa terdapat korelasi positif antara  pada sektor ekonomi, pendidikan, dan kesehatan terhadap tingkat Indeks Pembangunan Manusia Indonesia. Semakin besar pemerintah membelanjakan uangnya pada sektor yang menunjang kegiatan ekonomi, pendidikan, dan kesehatan, maka tingkat Indeks Pembangunan Manusia juga akan semakin meningkat. Begitu juga sebaliknya, semakin sedikit pemerintah membelanjakan uangnya pada sektor ekonomi, pendidikan, dan kesehatan, maka Indeks Pembangunan Manusia juga akan menurun.

# One to many relationship

## 1. Report the account representative for each customer.

```python
# Termasuk customer yang belum memiliki account representative agar lebih informatif
run("""
    SELECT
    customerNumber,
    customerName,
    employeeNumber,
    CONCAT(firstName, lastName)
    FROM
    customers c
    LEFT JOIN 
    employees e
    ON c.salesRepEmployeeNumber = e.employeeNumber
    ORDER BY employeeNumber DESC
""")
```

## 2. Report total payments for Atelier graphique.

```python
run("""
    SELECT
    c.customerNumber,
    customerName,
    SUM(amount)
    FROM
    customers c
    INNER JOIN
    payments p
    ON c.customerNumber = p.customerNumber
    WHERE c.customerNumber = '103'
    
""")
```

## 3. Report the total payments by date

```python
run("""
    SELECT
    paymentDate,
    SUM(amount)
    FROM
    payments
    GROUP BY paymentDate
    ORDER BY paymentDate
""")
```

## 4. Report the products that have not been sold.

```python
run("""
    SELECT DISTINCT
    p.productCode,
    p.productName,
    o.orderNumber
    FROM
    products p
    LEFT JOIN
    orderdetails o
    ON 
    p.productCode = o.productCode
    WHERE o.orderNumber is null
""")
```

## 5. List the amount paid by each customer.

```python
run("""
    SELECT
    c.customerNumber,
    c.customerName,
    SUM(p.amount)
    FROM
    customers c
    INNER JOIN
    payments p
    ON
    c.customerNumber = p.customerNumber
    GROUP BY c.customerNumber
""")
```

## 6. How many orders have been placed by Herkku Gifts?

```python
run("""
    SELECT
    c.customerNumber,
    c.customerName,
    count(o.orderNumber)
    FROM
    customers c
    INNER JOIN
    orders o
    ON
    c.customerNumber = o.customerNumber
    WHERE c.customerName = 'Herkku Gifts'
    GROUP BY c.customerNumber
""")
```

## 7. Who are the employees in Boston?

```python
run("""
    SELECT
    e.employeeNumber,
    CONCAT(e.firstName, e.lastName)
    FROM employees e
    INNER JOIN
    offices o
    ON
    e.officeCode = o.officeCode
    WHERE o.city = 'Boston'
""")
```

## 8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
    SELECT
    c.customerNumber,
    c.customerName,
    SUM(p.amount) Total_Payment
    FROM
    customers c
    INNER JOIN
    payments p
    ON
    c.customerNumber = p.customerNumber
    GROUP BY c.customerNumber
    HAVING Total_Payment > 100000
    ORDER BY Total_Payment DESC
""")
```

## 9. List the value of 'On Hold' orders.

```python
run("""
    SELECT DISTINCT
    o.orderNumber,
    o.orderDate,
    o.requiredDate,
    o.status,
    o.comments,
    o.customerNumber,
    SUM(od.quantityOrdered*od.priceEach) Value_Of_On_Hold_Orders
    FROM 
    orders o
    INNER JOIN
    orderdetails od
    ON
    o.orderNumber = od.orderNumber
    WHERE o.status = 'on hold'
    GROUP BY o.orderNumber
""")
```

## 10. Report the number of orders 'On Hold' for each customer.

```python
run("""
    SELECT
    c.customerNumber,
    c.customerName,
    COUNT(CASE WHEN o.status = 'On Hold' THEN 1 END) On_Hold_Orders
    FROM
    customers c
    LEFT JOIN
    orders o
    ON
    c.customerNumber = o.customerNumber
    GROUP BY c.customerNumber
    ORDER BY On_Hold_Orders DESC, c.customerNumber
""")
```

# Many to many relationship

## 1. List products sold by order date.

```python
run("""
    SELECT
    o.orderDate,
    od.productCode,
    p.productName
    FROM
    orders o
    INNER JOIN
    orderdetails od
    ON
    o.orderNumber = od.orderNumber
    INNER JOIN
    products p
    ON
    od.productCode = p.productCode
    ORDER BY o.orderDate
""")
```

## 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
    SELECT
    o.orderDate,
    o.orderNumber,
    od.productCode,
    p.productName
    FROM
    orders o
    INNER JOIN
    orderdetails od
    ON
    o.orderNumber = od.orderNumber
    INNER JOIN
    products p
    ON
    od.productCode = p.productCode
    WHERE
    p.productName = '1940 Ford Pickup Truck'
    ORDER BY o.orderDate DESC
""")
```

## 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
    SELECT
    c.customerNumber,
    c.customerName,
    o.orderNumber,
    SUM(od.quantityOrdered*od.priceEach) as Value
    FROM
    customers c
    INNER JOIN
    orders o
    ON
    c.customerNumber = o.customerNumber
    INNER JOIN
    orderdetails od
    ON
    o.orderNumber = od.orderNumber
    GROUP BY od.orderNumber
    HAVING Value > 25000
""")
```

## 5. List the names of products sold at less than 80% of the MSRP.

```python
run("""
    SELECT
    productName as productName_sold_less_than80percentMSRP,
    (od.priceEach/p.MSRP)*100 as Price_Percentage_from_MSRP,
    orderNumber as in_this_orderNumber
    FROM
    products p
    INNER JOIN
    orderdetails od
    ON
    p.productCode = od.productCode
    HAVING Price_Percentage_from_MSRP < 80
""")
```

## 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
    SELECT
    od.productCode,
    p.productName,
    priceEach,
    (priceEach-buyPrice)/buyPrice*100 as Percentage_Margin,
    orderNumber
    FROM
    orderdetails od
    INNER JOIN
    products p
    ON od.productCode = p.productCode
    Having Percentage_Margin >= 100 
    ORDER BY productCode, PriceEach, orderNumber
""")
```

## 7. List the products ordered on a Monday.

```python
run("""
    SELECT
    p.productName,
    o.orderDate,
    DAYNAME(o.orderDate) as Day
    FROM
    orders o
    INNER JOIN
    orderdetails od
    ON
    o.orderNumber = od.orderNumber
    INNER JOIN
    products p
    ON
    od.productCode = p.productCode
    HAVING Day = 'Monday'
""")
```

## 8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
    SELECT DISTINCT
    od.productCode,
    o.status,
    SUM(od.quantityOrdered) as quantity_on_hand
    FROM 
    orders o
    INNER JOIN
    orderdetails od
    ON
    o.orderNumber = od.orderNumber
    WHERE o.status = 'on hold'
    GROUP BY od.productCode
    ORDER BY productCode
""")
```
